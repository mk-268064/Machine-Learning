# Importing necessary modules for **linear regression**
--pandas,numpy,train_test_split,LinearRegression--

**train_test_split**: To split dataset for training and testing.
**LinearRegression**: Implements Linear regression.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

# Explore **Dataset**

```python
dataset = pd.read_csv('/content/Salary_data.csv')
print(dataset.head())
print(dataset.isnull().sum())
dataset = dataset.dropna()
```

#Select **features(x)** and **Target(y)**:


```python
x = dataset.iloc[:,[0,-2]].values
print(x)
```

```python
y = dataset.iloc[:,-1].values
print(y)
```

#**Pandas .iloc[rows,columns]**
.values convert dataframe into numpy values

#***Split Dataset into Training dataset***


1.   70% data is used for testing(test_size=0.3 means 30%)
2.   random_state=42 ensures split is reproducible




```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

#***Training the Linear Regression Model***


*   The model learns the best m (slope) and c (intercept) for the equation
y=mX+c.




```python
model = LinearRegression()
model.fit(x_train,y_train)
```

# ***Make Predictions***
Model predicts based on the test set

```python
y_pred = model.predict(x_test)
```

```python
plt.scatter(x_train[:,0],y_train,color='red')
plt.plot(x_train[:,0],model.predict(x_train),color='blue')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
